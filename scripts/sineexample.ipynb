# Example of Cairo plot

```julia
include("CairoPlot.jl")
```

```julia
Nx = 640
Ny = 384
title = "Perturbation of a sine wave!\nwhat else?"
x = range(1, 30, length=30)
y = sin.(x/5) .+ 0.4randn(length(x))
xticks = 1:5:30
yticks = -1.5:0.5:1.5
```

```julia
c = crplot(x, y; xticks, yticks, Nx, Ny, title, padding=(0.08, 0.02, 0.06, 0.08))
```

```julia
display(c)
```

```julia
write_to_png(c,"cairoplotsineexample.png")
```
